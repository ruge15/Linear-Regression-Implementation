### Now putting this values into features and labels for further model building

### Dividing the data into train and test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(features,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=10)
```

# Building Linear Regression Model

```python
#Import the model from sklearn
from sklearn.linear_model import LinearRegression
```

 First building the model without Standardizing the data

```python
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
#
# Check the quality of the model
#
# SL(significance Level) ------ Standard Alpha Values --- (0.05, 0.01, 0.1)
#
# SL = 0.05
# CL = 1 - SL = 0.95 (The project is expecting atleast 95% accuracy in prediction)
#
# When it comes to approving a model, always go for GENERALIZED MODEL
#
# Generalized model means something that satisfies the following criteria
#
# a. Calc the evaluation score of the model for both training data and testing data
#         trainScore (Score generated by training data)
#         testScore  (Score generated by testing data)
#
#   Criteria is
#     if testScore > trainScore and testScore >= CL:
#                    Approve the model
#     else:
#                    Reject the model and start the process of improving the model
#
```

```python
trainScore=model.score(X_train, y_train)
testScore=model.score(X_test, y_test)

print(f"Train Score is :{trainScore} and Test Score is : {testScore}")
```

```python
# Lets check for the prediction and plot the same

model_predict=model.predict(X_test)

#Lets see the difference between the predicted and actuall values using the KDE plot

sns.displot(model_predict-y_test, kind='kde')
```

```python
#Lets check for the R2 score

from sklearn.metrics import r2_score

score=r2_score(model_predict, y_test)

print(score)
```

```python
## Lets see if different Random State gives better answer

CL=0.6

for seed in range (1,100):
    X_train, X_test, y_train, y_test= train_test_split(features,
                                                      label,
                                                      test_size=0.2,
                                                      random_state=seed)


    model2=LinearRegression()
    model2.fit(X_train, y_train)

    trainScore2=model2.score(X_train, y_train)
    testScore2=model2.score(X_test, y_test)

    if testScore2 > trainScore2 and testScore2 >= CL:
        print(f"Test Score: {testScore2} and Train Score : {testScore2} for Random State: {seed}")
```

# We can conclude with this that this model does not cross the CL level more than 0.7 and need more improvment in the model, which will be done as and when we progress
